```python
# Librerías7

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
```

```python
Path = "mapache.png"

with Image.open(Path, 'r') as im:
    pixel_array = np.array(im, dtype="uint8")
    
print(pixel_array)
```

```python
dt_mapache = pd.read_csv("mapache.csv", delimiter=";")
dt_mapache
```

```python
im = Image.fromarray(np.array(dt_mapache, dtype="uint8"), mode="L")
im.resize((256,256))
```

```python
## Filtros
# Utilizando la función kernel

identidad = [0,0,0,
             0,1,0,
             0,0,0]

k = ImageFilter.Kernel(size=(3,3), kernel = identidad, scale = 1)

im2 = im.filter(k)
im2
```

```python
# Desenfoque

desenfoque = [1 for i in range(9)]

k = ImageFilter.Kernel(size=(3,3), kernel=desenfoque, scale=9)
im2 = im.filter(k)
im2
```

```python
# Enfoque
enfoque = [0,-1,0,
           -1,5,-1,
           0,-1,0]
k = ImageFilter.Kernel(size=(3,3), kernel=enfoque, scale=1)
im2 = im.filter(k)
im2
```

```python
# Emboss
eemboss = [-2,-1,0,
           -1,1,1,
           0,1,2]
k = ImageFilter.Kernel(size=(3,3), kernel=eemboss, scale=1)
im2 = im.filter(k)
im2
```

```python
# sobel horizontal
sobel = [-1,-2,-1,
           0,0,0,
           1,2,1]
k = ImageFilter.Kernel(size=(3,3), kernel=sobel, scale=1)
im2 = im.filter(k)
im2
```
