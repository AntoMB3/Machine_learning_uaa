# Práctica No. 3 Visualización de datos

```python
# Librerías
import matplotlib.pyplot as plt
import numpy as np
```

### Gráfico de funciones

```python
# Datos del eje X para los siguientes gráficos
x = np.linspace(0,10,100) # 100 datos de 0 a 10
print(x)
```

#### Lo siguiente se ejecuta junto dentro del mismo gráfico

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

#### Paneles de gráficos

```python
plt.figure() # Crear la figura

# Crear primer panel
plt.subplot(2,1,1) #(filas, columnas, #numero de panel)
plt.plot(x,np.sin(x))

#Crear el segundo panel
plt.subplot(2,1,2) #(filas, columnas, #numero de panel)
plt.plot(x,np.cos(x))
```

#### Un estilo orientado a objetos para situaciones más complejas

```python
fig, ax = plt.subplots(2)

# Llama el método plot() method

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

##### Usando colores

```python
plt.plot(x, np.sin(x - 0), color='blue') # Color
plt.plot(x, np.sin(x - 1), color='g')  # Clave de color
plt.plot(x, np.sin(x - 2), color='0.75') #Escala de grises
plt.plot(x, np.sin(x - 3), color='#FFDD44') #Codigo Hexadecimal
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) #Tupla RGB
plt.plot(x, np.sin(x - 5), color='chartreuse') # Nombres de color en HTML
```

```python
## Usando lineas puntadas
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

### Cambiamos los límites de los ejes

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

### Títulos

```python
plt.plot(x, np.sin(x))
plt.title("Funcion Sen(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

### Leyendas

```python
plt.plot(x, np.sin(x), '-g', label="Seno(x)")
plt.plot(x, np.cos(x), ':b', label="Cos(x)")
plt.legend(loc="upper right")
```

## Ejemplos con datos de Iris

```python
import numpy as np
import pandas as pd

df = pd.read_csv("iris.csv", delimiter=";",decimal=".")
print(df.shape)
print(df.head())
```

## Eliminar las variables categóricas

```python
x = df.iloc[0:100, [0,2]].values
```

```python
import matplotlib.pyplot as plt
#Graficamos datos
plt.scatter(x[:50,0], x[:50,1],
            color="orange", marker="o", label="Setosa")
plt.scatter(x[50:100,0], x[50:100,1],
            color="blue", marker="x", label="Versicolor")

#Leyenda
plt.legend(loc="upper left")

# Graficamos etiquetas
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris)

caracteristicas = iris.data.T
print(caracteristicas)

plt.scatter(caracteristicas[0], caracteristicas[1],
            alpha=0.2,s=100*caracteristicas[3], c=iris.target, cmap="viridis")
plt.xlabel(iris.feature_names[0])
plt.xlabel(iris.feature_names[1])
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
print(iris.head())
corr = sns.pairplot(iris, hue="species", size=2.5)
```
