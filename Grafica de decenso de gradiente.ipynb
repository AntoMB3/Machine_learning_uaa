```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funcion
def fun(x):
    return (2*(x**2)*np.cos(x)) - 5*x
```

```python
# Valores de x de [-5,5]
x = np.linspace(-5,5,50)
x
```

```python
# Plotear la función

plt.plot(x, fun(x), color="black")
```

```python
# Funcion de derivada
def func_derivada(x):
    return -2*(x**2)*np.sin(x) + 4 * x * np.cos(x) - 5
```

```python
# Razond e aprendizaje
n = 0.05
```

```python
# Punto random en x
x0 = np.random.randint(0,10) - 5
x0


y = func_derivada(x0)
#Graficamos el punto
plt.scatter(x0, fun(x0), color="red")

x1 = x0 - n*y

# Plotear la función

plt.plot(x, fun(x), color="black")

```

```python
# Lo realizamos de manera iterativa
# Punto random en x
x0 = np.random.randint(0,10) - 5
x0
for i in range(50):
    
    y = func_derivada(x0)
    #Graficamos el punto
    plt.scatter(x0, fun(x0), color="red")

    x1 = x0 - n*y
    x0 = x1

# Plotear la función

plt.plot(x, fun(x), color="black")
```
