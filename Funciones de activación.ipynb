# Funciones de activaci贸n

Concepto: Las funciones de activaci贸n se utilizan para propagar la salida hacia adelante. La salida las reciben las siguientes capas hasta la salida. Sirven para implementar la no linealidad de la red neuronal.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Activaci贸n Binaria

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)

x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function :binaryStep')
plt.show()
```

### Activaci贸n tangencial

```python
def tanh(x):
    ''' It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.'''

    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

### Activaci贸n sigmoidea

Siempre da valores reales cercanos entre 0 y 1, entre mayor es el valor de entrada se aproxima a 1 el valor, si es un n煤mero negativo se aproxima a 0.

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

### Softmax

```python
def softmax(x):
    ''' Compute softmax values for each sets of scores in x. '''
    return np.exp(x) / np.sum(np.exp(x))

x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```
