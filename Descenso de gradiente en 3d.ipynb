## Descenso de gradiente 3D

```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
import pandas as pd
```

```python
def f(x,y):
    return 3 * x * x + 4 * y * y

def f_x(x,y):
    return 3 * x

def f_y(x,y):
    return 4 * y
```

```python
x = np.array(np.arange(-10,10,1))
y = np.array(x.copy().T)

z = f(x,y)

data = pd.DataFrame({'x': x, 'y': y, 'z':z})
data

x.ndim
```

```python

import plotly.graph_objects as go 
import numpy as np 
  
x = np.outer(np.linspace(-10, 10, 30), np.ones(30)) 
y = x.copy().T 
z = f(x,y)
  
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])

x0 = np.random.randint(0,10) - 5
y0 = np.random.randint(0,10) - 5

n = 0.05

for i in range(50):
    
    z = f(x0,y0)

    fig.add_trace(
        go.Scatter3d(x=[x0],
                    y=[y0],
                    z=[z],
                    mode='markers'
                    )
    )
    
    x1 = x0 - n*f_x(x0,y0)
    y1 = y0 - n*f_y(x0,y0)
    x0 = x1
    y0 = y1
  
fig.show() 
```
